```python
import os
import pandas as pd
import json
import mysql.connector
```

```python
# Aggregate Transaction

path="E:/VScode/Phone Pe Pulse/pulse/data/aggregated/transaction/country/india/state/"
agg_trans_list=os.listdir(path)

agg_trans_dict={'State':[] , 'Year':[] , 'Quater':[] , 'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in agg_trans_list:
    path_i=path+i+"/"
    agg_trans_year=os.listdir(path_i)
    for j in agg_trans_year:
        path_j=path_i+j+"/"
        agg_trans_json=os.listdir(path_j)
        for k in agg_trans_json:
            path_k=path_j+k
            data=open(path_k,'r')
            d=json.load(data)
            for values in d['data']['transactionData']:
                count=values['paymentInstruments'][0]['count']
                amount=values['paymentInstruments'][0]['amount']
                agg_trans_dict['Transaction_type'].append(values['name'])
                agg_trans_dict['Transaction_count'].append(count)
                agg_trans_dict['Transaction_amount'].append(amount)
                agg_trans_dict['State'].append(i)
                agg_trans_dict['Year'].append(j)
                agg_trans_dict['Quater'].append(int(k.strip(".json")))

agg_trans_df=pd.DataFrame(agg_trans_dict)
            
agg_trans_df['State']=agg_trans_df['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
agg_trans_df['State']=agg_trans_df['State'].str.replace("-"," ")
agg_trans_df['State']=agg_trans_df['State'].str.title()
agg_trans_df['State']=agg_trans_df['State'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```

```python
# Aggregate User

path="E:/VScode/Phone Pe Pulse/pulse/data/aggregated/user/country/india/state/"
agg_user_list=os.listdir(path)

agg_user_dict={'State':[] , 'Year':[] , 'Quater':[] , 'Brands':[], 'Transaction_count':[], 'Percentage':[]}

for i in agg_user_list:
    path_i=path+i+"/"
    agg_user_year=os.listdir(path_i)
    for j in agg_user_year:
        path_j=path_i+j+"/"
        agg_user_json=os.listdir(path_j)
        for k in agg_user_json:
            path_k=path_j+k
            data=open(path_k,'r')
            d1=json.load(data)
            try:
                for values in d1['data']['usersByDevice']:
                    brand=values['brand']
                    count=values['count']
                    percentage=values['percentage']
                    agg_user_dict['Brands'].append(brand)
                    agg_user_dict['Transaction_count'].append(count)
                    agg_user_dict['Percentage'].append(percentage)
                    agg_user_dict['State'].append(i)
                    agg_user_dict['Year'].append(j)
                    agg_user_dict['Quater'].append(int(k.strip(".json")))
            except:
                pass

agg_user_df=pd.DataFrame(agg_user_dict)

agg_user_df['State']=agg_user_df['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
agg_user_df['State']=agg_user_df['State'].str.replace("-"," ")
agg_user_df['State']=agg_user_df['State'].str.title()
agg_user_df['State']=agg_user_df['State'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')


```

```python
# Map Transaction

path="E:/VScode/Phone Pe Pulse/pulse/data/map/transaction/hover/country/india/state/"
map_trans_list=os.listdir(path)

map_trans_dict={'State':[] , 'Year':[] , 'Quater':[] , 'District':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in map_trans_list:
    path_i=path+i+"/"
    map_trans_year=os.listdir(path_i)
    for j in map_trans_year:
        path_j=path_i+j+"/"
        map_trans_json=os.listdir(path_j)
        for k in map_trans_json:
            path_k=path_j+k
            data=open(path_k,'r')
            d2=json.load(data)
            for values in d2['data']['hoverDataList']:
                name=values['name']
                count=values['metric'][0]['count']
                amount=values['metric'][0]['amount']
                map_trans_dict['District'].append(name)
                map_trans_dict['Transaction_count'].append(count)
                map_trans_dict['Transaction_amount'].append(amount)
                map_trans_dict['State'].append(i)
                map_trans_dict['Year'].append(j)
                map_trans_dict['Quater'].append(int(k.strip(".json")))
            
map_trans_df=pd.DataFrame(map_trans_dict)

map_trans_df['State']=map_trans_df['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
map_trans_df['State']=map_trans_df['State'].str.replace("-"," ")
map_trans_df['State']=map_trans_df['State'].str.title()
map_trans_df['State']=map_trans_df['State'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
map_trans_df['District']=map_trans_df['District'].str.replace(' district','')
map_trans_df['District']=map_trans_df['District'].str.title()
```

```python
# Map User

path="E:/VScode/Phone Pe Pulse/pulse/data/map/user/hover/country/india/state/"
map_user_list=os.listdir(path)

map_user_dict={'State':[] , 'Year':[] , 'Quater':[] , 'District':[], 'Registered_users':[], 'App_opens':[]}

for i in map_user_list:
    path_i=path+i+"/"
    map_user_year=os.listdir(path_i)
    for j in map_user_year:
        path_j=path_i+j+"/"
        map_user_json=os.listdir(path_j)
        for k in map_user_json:
            path_k=path_j+k
            data=open(path_k,'r')
            d3=json.load(data)
            for values in d3['data']['hoverData'].items():
                register_users=values[1]['registeredUsers']
                app_opens=values[1]['appOpens']
                map_user_dict['District'].append(values[0])
                map_user_dict['Registered_users'].append(register_users)
                map_user_dict['App_opens'].append(app_opens)
                map_user_dict['State'].append(i)
                map_user_dict['Year'].append(j)
                map_user_dict['Quater'].append(int(k.strip(".json")))

map_user_df=pd.DataFrame(map_user_dict)

map_user_df['State']=map_user_df['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
map_user_df['State']=map_user_df['State'].str.replace("-"," ")
map_user_df['State']=map_user_df['State'].str.title()
map_user_df['State']=map_user_df['State'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
map_user_df['District']=map_user_df['District'].str.replace(' district','')
map_user_df['District']=map_user_df['District'].str.title()
```
